```python
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris = load_iris()
data, target = iris.data, iris.target
```

```python
it=iter(data)
```

```python
for i in it:
    print(i)
```

```python
try:
    print(next(it))
except StopIteration:
    print("The iterator is empty")
```

```python
##generators
def gen(data,target):
    for i in range(len(data)):
        yield data[i],target[i]
```

```python
g=gen(data,target)
```

```python
for i in g:
    print(i)
```

```python
##decorators
def dec(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result
    return wrapper
@dec
def iris(data, target):
    processed_data = [sum(row) for row in data]
    result = {"processed_data": processed_data, "target": target}
    return result
result = iris(data, target)
print(result)
```
