```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
df=pd.DataFrame(digits.data,digits.target)
df
```

```python
df['target']=digits.target
df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
linear_regressor = LinearRegression(fit_intercept=True, copy_X=True,n_jobs=None,positive=False)
```

```python
linear_regressor.fit(X_train,y_train)
```

```python
print(linear_regressor.score(X_train,y_train))
print(linear_regressor.score(X_test,y_test))
```

```python
y_pred=linear_regressor.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.svm import SVC
```

```python
model=SVC(C=20,gamma='scale',kernel='rbf')
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimators=[int(x) for x in np.linspace(start=10,stop=80,num=10)]
max_features=['auto','sqrt','log2']
max_depth=[1,2,3,4]
min_samples_split=[1,2,5,10,20]
min_samples_leaf=[1,2,3,4]
bootstrap=[True,False]
```

```python
param_grid={
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
    'bootstrap':bootstrap
}
print(param_grid)
```

```python
rf_model=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=rf_model,param_grid=param_grid,cv=2,verbose=2,n_jobs=4)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
print(f'train accuracy: {grid_search.score(X_train,y_train):.3f}')
print(f'test accuracy: {grid_search.score(X_test,y_test):.3f}')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_reg=LogisticRegression()
```

```python
param_grid={
    'penalty':['l1','l2'],
    'C':[0.001,0.01,0.1,1,10],
    'solver':['liblinear','lbfgs','saga']
}
```

```python
grid_search=GridSearchCV(estimator=logistic_reg,param_grid=param_grid,cv=3,verbose=1,n_jobs=-1)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
print(grid_search.score(X_train,y_train))
print(grid_search.score(X_test,y_test))
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
accuracy = (y_pred == y_test).mean()
print("Accuracy: ", accuracy)
```

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt_classifier = DecisionTreeClassifier()
```

```python
grid_search = GridSearchCV(estimator=dt_classifier,param_grid=param_grid,cv=3,n_jobs=-1)
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.score(X_train,y_train))
print(grid_search.score(X_test,y_test))
```

```python
best_params=grid_search.best_params_
best_dt_classifier = DecisionTreeClassifier(**best_params)
best_dt_classifier.fit(X_train, y_train)
```

```python
y_pred = best_dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```
